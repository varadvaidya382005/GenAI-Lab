```python
# Cell 2: Create Pipeline
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
pipeline = pipeline.to("cuda" if torch.cuda.is_available() else "cpu")

print("✓ Text-to-Image pipeline ready")
```

```python
# Cell 3: Advanced Generation with Parameters
def create_image(
    prompt,
    width=512,
    height=512,
    num_images=1,
    guidance_scale=7.5
):
    images = pipeline(
        prompt,
        width=width,
        height=height,
        num_images_per_prompt=num_images,
        guidance_scale=guidance_scale
    ).images
    return images

print("✓ Image creation function ready")
```

```python
# Cell 4: Batch Generation
prompts = [
    "A professional product photo of a smartwatch",
    "An architectural rendering of a modern house",
    "A logo design for a tech startup",
]

for idx, prompt in enumerate(prompts):
    print(f"\nGenerating {idx+1}/{len(prompts)}: {prompt}")
    images = create_image(prompt, num_images=3)
    for img_idx, img in enumerate(images):
        filename = f"generated_{idx}_{img_idx}.png"
        img.save(filename)
        print(f"  ✓ Saved {filename}")

print("\n✓ All images generated!")
```
