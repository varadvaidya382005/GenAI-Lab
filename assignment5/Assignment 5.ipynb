```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

if torch.cuda.is_available():
    pipe = StableDiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16
    ).to("cuda")
else:
    pipe = StableDiffusionPipeline.from_pretrained(
        model_id
    ).to("cpu")

print("Device:", pipe.device)
```

```python
prompt = "A fantasy girl holding a glowing crystal orb, magical forest, butterflies, ultra detailed, cinematic lighting"

image = pipe(
    prompt=prompt,
    guidance_scale=7.5,
    num_inference_steps=30
).images[0]

image.save("creative_artwork.png")
image
```
